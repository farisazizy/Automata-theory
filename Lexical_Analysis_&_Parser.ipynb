```python

def Lexical(kata):
    import string

    alphabet = list(string.ascii_lowercase)
    state_int = list(range(1, 51))
    stateList = [str(int) for int in state_int]

    transisi = {}

    for state in stateList:
        for j in alphabet:
            transisi[(state, j)] = 'error'
        transisi[(state, '#')] = 'error'    
        transisi[(state, ' ')] = 'error'
    
    #state awal
    transisi[('1', ' ')] = '1'

    #SB
    #abdi
    transisi[('1', 'a')] = '2'
    transisi[('2', 'b')] = '9'
    transisi[('9', 'd')] = '18'
    transisi[('18', 'i')] = '49'

    #bapa
    transisi[('1', 'b')] = '3'
    transisi[('3', 'a')] = '10'
    transisi[('10', 'p')] = '19'
    transisi[('19', 'a')] = '49'

    #lanceuk
    transisi[('1', 'l')] = '6'
    transisi[('6', 'a')] = '14'
    transisi[('14', 'n')] = '21'
    transisi[('21', 'c')] = '28'
    transisi[('28', 'e')] = '34'
    transisi[('34', 'u')] = '41'
    transisi[('41', 'k')] = '49'

    #indung
    transisi[('1', 'i')] = '5'
    transisi[('5', 'n')] = '13'
    transisi[('13', 'd')] = '20'
    transisi[('20', 'u')] = '27'
    transisi[('27', 'n')] = '33'
    transisi[('33', 'g')] = '49'

    #VB
    #ngadahar
    transisi[('1', 'n')] = '8'
    transisi[('8', 'g')] = '16'
    transisi[('16', 'a')] = '23'
    transisi[('23', 'd')] = '29'
    transisi[('29', 'a')] = '35'
    transisi[('35', 'h')] = '40'
    transisi[('40', 'a')] = '45'
    transisi[('45', 'r')] = '49'

    #motong
    transisi[('1', 'm')] = '7'
    transisi[('7', 'o')] = '15'
    transisi[('15', 't')] = '22'
    transisi[('22', 'o')] = '27'
    transisi[('27', 'n')] = '33'
    transisi[('33', 'g')] = '49'

    #nginum
    transisi[('1', 'n')] = '8'
    transisi[('8', 'g')] = '16'
    transisi[('16', 'i')] = '36'
    transisi[('36', 'n')] = '42'
    transisi[('42', 'u')] = '46'
    transisi[('46', 'm')] = '49'

    #nyandak
    transisi[('1', 'n')] = '8'
    transisi[('8', 'y')] = '17'
    transisi[('17', 'a')] = '24'
    transisi[('24', 'n')] = '30'
    transisi[('30', 'd')] = '37'
    transisi[('37', 'a')] = '41'
    transisi[('41', 'k')] = '49'
    
    #OB
    #cai
    transisi[('1', 'c')] = '11'
    transisi[('11', 'a')] = '18'
    transisi[('18', 'i')] = '49'

    #hawa
    transisi[('1', 'h')] = '4'
    transisi[('4', 'a')] = '12'
    transisi[('12', 'w')] = '19'
    transisi[('19', 'a')] = '49'

    #seuneu
    transisi[('1', 's')] = '25'
    transisi[('25', 'e')] = '31'
    transisi[('31', 'u')] = '38'
    transisi[('38', 'n')] = '43'
    transisi[('43', 'e')] = '47'
    transisi[('47', 'u')] = '49'

    #taneuh
    transisi[('1', 't')] = '26'
    transisi[('26', 'a')] = '32'
    transisi[('32', 'n')] = '39'
    transisi[('39', 'e')] = '44'
    transisi[('44', 'u')] = '48'
    transisi[('48', 'h')] = '49'

    transisi[('49', '#')] = 'accept'
    transisi[('49', ' ')] = '50'
    transisi[('50', '#')] = 'accept'
    transisi[('50', ' ')] = '50'

    transisi[('50', 'h')] = '4'
    transisi[('50', 'b')] = '3'
    transisi[('50', 'a')] = '2'
    transisi[('50', 'i')] = '5'
    transisi[('50', 'l')] = '6'
    transisi[('50', 'm')] = '7'
    transisi[('50', 'n')] = '8'
    transisi[('50', 's')] = '25'
    transisi[('50', 't')] = '26'
    transisi[('50', 'c')] = '11'

    index = 0
    state = '1'
    curr_token = ''
    temp_kata = ''
    inputString = kata.lower() + '#'

    while state != 'accept':
        curr_char = inputString[index]
        curr_token += curr_char
        temp_kata += curr_char
        state = transisi[(state, curr_char)]
        if state == '49':
            curr_token = ''
        if state == 'error':
            return False
            break
        index = index + 1
    
    if state == 'accept':
        return True
    else:
        return False

def Parser_(kata):
    tokens = kata.lower().split()
    tokens.append('EOS')

    non_terminals = ['S', 'SB', 'VB', 'OB']
    terminals = ['bapa', 'indung', 'lanceuk', 'abdi', 'ngadahar', 'motong', 
                 'nginum', 'nyandak', 'cai','hawa', 'seuneu', 'taneuh']

    parser = {}

    parser[('S', 'bapa')] = ['SB', 'VB', 'OB']
    parser[('S', 'indung')] = ['SB', 'VB', 'OB']
    parser[('S', 'lanceuk')] = ['SB', 'VB', 'OB']
    parser[('S', 'abdi')] = ['SB', 'VB', 'OB']
    parser[('S', 'ngadahar')] = [('error')]
    parser[('S', 'motong')] = [('error')]
    parser[('S', 'nginum')] = [('error')]
    parser[('S', 'nyandak')] = [('error')]
    parser[('S', 'cai')] = [('error')]
    parser[('S', 'hawa')] = [('error')]
    parser[('S', 'seuneu')] = [('error')]
    parser[('S', 'taneuh')] = [('error')]
    parser[('S', 'EOS')] = [('error')]
    
    parser[('SB', 'bapa')] = [('bapa')]
    parser[('SB', 'indung')] = [('indung')]
    parser[('SB', 'lanceuk')] = [('lanceuk')]
    parser[('SB', 'abdi')] = [('abdi')]
    parser[('SB', 'ngadahar')] = [('error')]
    parser[('SB', 'motong')] = [('error')]
    parser[('SB', 'nginum')] = [('error')]
    parser[('SB', 'nyandak')] = [('error')]
    parser[('SB', 'cai')] = [('error')]
    parser[('SB', 'hawa')] = [('error')]
    parser[('SB', 'seuneu')] = [('error')]
    parser[('SB', 'taneuh')] = [('error')]
    parser[('SB', 'EOS')] = [('error')]

    parser[('VB', 'bapa')] = [('error')]
    parser[('VB', 'indung')] = [('error')]
    parser[('VB', 'lanceuk')] = [('error')]
    parser[('VB', 'abdi')] = [('error')]
    parser[('VB', 'ngadahar')] = [('ngadahar')]
    parser[('VB', 'motong')] = [('motong')]
    parser[('VB', 'nginum')] = [('nginum')]
    parser[('VB', 'nyandak')] = [('nyandak')]
    parser[('VB', 'cai')] = [('error')]
    parser[('VB', 'hawa')] = [('error')]
    parser[('VB', 'seuneu')] = [('error')]
    parser[('VB', 'taneuh')] = [('error')]
    parser[('VB', 'EOS')] = [('error')]

    parser[('OB', 'bapa')] = [('error')]
    parser[('OB', 'indung')] = [('error')]
    parser[('OB', 'lanceuk')] = [('error')]
    parser[('OB', 'abdi')] = [('error')]
    parser[('OB', 'ngadahar')] = [('error')]
    parser[('OB', 'motong')] = [('error')]
    parser[('OB', 'nginum')] = [('error')]
    parser[('OB', 'nyandak')] = [('error')]
    parser[('OB', 'cai')] = [('cai')]
    parser[('OB', 'hawa')] = [('hawa')]
    parser[('OB', 'seuneu')] = [('seuneu')]
    parser[('OB', 'taneuh')] = [('taneuh')]
    parser[('OB', 'EOS')] = [('error')]

    stack = []
    stack.append('#')
    stack.append('S')

    index = 0
    symbol = tokens[index]
    top = stack[len(stack) - 1]

    while (len(stack) > 0):
        top = stack[len(stack) - 1]
        #print('top : ', top)
        #print('symbol : ', symbol)
        if top in terminals:
            if top == symbol:
                stack.pop()
                index = index + 1
                symbol = tokens[index]
                if symbol == 'EOS':
                    stack.pop()
            else:
                #print('Error')
                return 'Error'
                break
        elif top in non_terminals:
            if parser[(top, symbol)][0] != 'error':
                stack.pop()
                symbols_pushed = parser[(top, symbol)]
                for i in range(len(symbols_pushed)-1, -1, -1):
                    stack.append(symbols_pushed[i])
            else:
                output = 'Kalimat ', kata, ' Tidak Valid, Tidak Sesuai Grammar'
                return "".join(output)
                break
        else:
            #print('error')
            return 'Error'
            break
        #print('isi stack : ', stack)
        #print()
        
    if symbol == 'EOS' and len(stack) == 0:
        #print('Input string ', kata, ', diterima sesuai grammer')
        output = 'Kalimat ', kata, ' Valid, Sesuai Grammar'
        return "".join(output)
    else:
        #print('error ', kata, ' tidak diterima')
        output = 'Kalimat ', kata, ' Tidak Valid, Tidak Sesuai Grammar'
        return "".join(output)

kata = input('Enter the sentence : ')

while kata != 'exit':
    lexicalCheck = Lexical(kata)
    if(lexicalCheck == True):
        print('Hasil Lexical Analysis : Valid')
        parserCheck = Parser_(kata)
        print("Hasil Parsing : ", parserCheck)
    else:
        print('Hasil Lexical Analysis : Tidak Valid')

    print()
    kata = input('Enter the sentence : ')





```
